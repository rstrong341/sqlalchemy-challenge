```python
#Import tools
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import datetime as dt

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func

```

```python
#database setup
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
#Reflect an existing database into a new model
Base = automap_base()

Base.prepare(engine, reflect = True)
```

```python
#Identify Classes
Base.classes.keys()
```

```python
#Create variables for the identified classes
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
#Link Python to the database
session = Session(engine)
```

```python
#Find most recent date
session.query(func.max(Measurement.date)).all()
```

```python
#Retrive the last 12 months of precipitation
twelve_months = dt.date(2017,8,23) - dt.timedelta(days=365)
```

```python
#Queey measurement data
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= twelve_months)
```

```python
#Put results into a dataframe, sort by date and plot
df=pd.DataFrame(results, columns=['date', 'Precipitation'])
df= df.set_index('date')
df = df.sort_index()

df.plot(rot=90)
plt.legend(loc='upper right')
plt.title('Precipitation between 2017-2018')
plt.xlabel('Date')
plt.ylabel('Precipitation')
```

```python
#Show the total number of stations
session.query(func.count(Station.station)).all()
```

```python
#Show the count of all the stations and order by count total
session.query(Measurement.station, func.count(Measurement.station)).\
    group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
```

```python
#Sort by min max and average
session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
    filter(Measurement.station == 'USC00519281').all()
```

```python
from pandas.plotting import table
#Retrive the last 12 months
twelve_months= dt.date(2017,8,23) - dt.timedelta(days=365)

#Gather results for the station with the highest count and filter for the twelve months
results = session.query(Measurement.tobs).\
    filter(Measurement.station == 'USC00519281').\
    filter(Measurement.date >= twelve_months).all()
df_tobs = pd.DataFrame(results, columns = ['tobs'])
#plot histogram and set titles/lables
df_tobs.plot.hist(bins=12)
plt.tight_layout()
plt.title('Temperature 2017-2018')
plt.xlabel('Date')
plt.ylabel('Temperature (farenheight)')

```
